```python
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
dataset.columns
```

```python
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
indep.shape
```

```python
dep=dataset[['classification_yes']]
dep.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(indep,dep,test_size=1/3,random_state=0)
```

```python
from sklearn.naive_bayes import CategoricalNB
from sklearn.model_selection import GridSearchCV
para={'alpha':[0.5,1.0,1.5,2.0,2.5,3.0],'fit_prior':[True,False]}
grid=GridSearchCV(CategoricalNB(),para,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(x_train,y_train)
```

```python
result=grid.cv_results_
table=pd.DataFrame.from_dict(result)
table
```

```python
table.to_excel("CKD_CategoricalNB_Grid.xlsx",index=False)
```

```python
y_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```

```python
from sklearn.metrics import f1_score
F1=f1_score(y_test,y_pred,average='weighted')
print(F1)
```

```python
from sklearn.metrics import roc_auc_score
score=roc_auc_score(y_test,y_pred)
print("The best parameter for CKD using CategoricalNB is: {}".format(grid.best_params_),score)
```
